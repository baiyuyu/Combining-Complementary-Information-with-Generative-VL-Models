```python
df_ground = pd.read_csv(file_path+'bd2bb_with_goldenwords.csv')
df_ground['golden_nouns']  = df_ground['golden_nouns'].apply(lambda x:eval(x))
df_ground.columns
```

```python
def cal_golden_words(sent,gold_words):
    num = 0
    try:
        for i in gold_words:
            if i in sent:
                num += 1
    except:
        print(sent,gold_words)
        
    return num
df.head()
```

```python
df  = pd.concat([df,df_ground[['golden_nouns', 'golden_nouns1', 'golden_nouns2',
       'golden_nouns3', 'golden_nouns4']]],axis=1)
df.columns
```
