```python
def formatanswer(x):
    result = x.split("Assistant: ")[-1]
#     if not x.startswith("I will"):
#         return "I will "+x
    if result.startswith("If"):
        result = x.split(', ')[-1]
    if result.startswith("you") or result.startswith("You"):
        result = result.lstrip(' ')
        result = (' ').join(['I','will'] + result.split(' ')[2:])
    return result



for i in range(5):
    df.loc[:,'answer_prompt_'+str(i)] = df.loc[:,'answer_prompt_'+str(i)].apply(lambda x:formatanswer(x))
```
